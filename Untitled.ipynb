```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_excel('data/netflix_dataset.xlsx')
df.head()
```

```python
def df_by_month(dataframe,month='2023-11'):
    return df.loc[df['week'].str.contains(month)]
```

```python
def genre_count(df):
    genre_count_dfs=[]
    for genre in ['genre_1','genre_2','genre_3']:
        genre_count_series=df[genre].value_counts()
        genre_count_df=pd.DataFrame({'genre':genre_count_series.index,'count':genre_count_series.values})
        genre_count_dfs.append(genre_count_df)
        
    genre_count_df=pd.concat(genre_count_dfs,ignore_index=True)    
    
    return genre_count_df.groupby('genre').mean().sort_values(by='count',ascending=False)
```

```python
def cast_member_count(df):
    cast_member_count_dfs=[]
    for cast_member in ['cast_member_1','cast_member_2','cast_member_3']:
        cast_member_count_series=df[cast_member].value_counts()
        cast_member_count_df=pd.DataFrame({'cast_member':cast_member_count_series.index,'count':cast_member_count_series.values})
        cast_member_count_dfs.append(cast_member_count_df)
        
    cast_member_count_df=pd.concat(cast_member_count_dfs,ignore_index=True)    
    
    return cast_member_count_df.groupby('cast_member').mean().sort_values(by='count',ascending=False)
```

```python
def mood_count(df):
    mood_count_dfs=[]
    for mood in ['mood_1','mood_2']:
        mood_count_series=df[mood].value_counts()
        mood_count_df=pd.DataFrame({'mood':mood_count_series.index,'count':mood_count_series.values})
        mood_count_dfs.append(mood_count_df)
        
    mood_count_df=pd.concat(mood_count_dfs,ignore_index=True)    
    
    return mood_count_df.groupby('mood').mean().sort_values(by='count',ascending=False)
```

```python
genre_count_month=genre_count(df_by_month(df,'2022-10')).head(10)

x=genre_count_month.index
y=genre_count_month.values.flatten()

plt.figure(figsize=(20,10))
sns.barplot(x=x,y=y)
```
